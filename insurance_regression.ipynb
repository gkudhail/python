```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
insurance_df = pd.read_csv('https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv')
```

```python

X = insurance_df.loc[::, 'age'].values
y = insurance_df.loc[::, 'charges'].values


plt.scatter(X, y, color='r')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = X.reshape(-1,1)
y = y.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X,y , test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

plt.scatter(X_test, y_test, color='b')
plt.plot(X_test, y_pred, color='m')

plt.xlabel('age')
plt.ylabel('Insurance Costs')

plt.show()
```

```python
print(f"The insurance cost someone likely to have if  40 : {model.predict([[40]])}")
```

```python
score = model.score(X_test, y_test)
score
```
